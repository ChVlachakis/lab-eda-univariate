### Part 2: Delving into Product Pricing

Business Question: How are products priced on Amazon UK, and are there specific price points or ranges that are more common?

#### Measures of Centrality:

- Calculate the mean, median, and mode for the price of products.
- What's the average price point of products listed? 
- How does this compare with the most common price point (mode)?

```python
mean_price = amazon_df["price"].mean()
median_price = amazon_df["price"].median()
median_price = amazon_df["price"].mode().iloc[0]

print(mean_price, median_price, median_price)   
```

The average price of the products listed is 89.24, which is much larger than the mode price which is 9.99. Aparently there might be product prices that are creating this big discrepancy.

#### Measures of Dispersion:

- Determine the variance, standard deviation, range, and interquartile range for product price.
- How varied are the product prices? 
- Are there any indicators of a significant spread in prices?

```python
var_price = amazon_df["price"].var()
std_price = amazon_df["price"].std()
range_price = amazon_df["price"].max()-amazon_df["price"].min()
iqr_price = amazon_df["price"].quantile(0.75) - amazon_df["price"].quantile(0.25)

var_price, std_price, range_price, iqr_price 
```

#### Visualizations:

- Is there a specific price range where most products fall? Plot a histogram to visualize the distribution of product prices. If its hard to read these diagrams, think why this is, and explain how it could be solved..
- Are there products that are priced significantly higher than the rest? Use a box plot to showcase the spread and potential outliers in product pricing.

```python
# plotting the histogram
import matplotlib.pyplot as plt

amazon_df["price"].hist(bins=50)
plt.xlabel("price")
plt.ylabel("frequency")
plt.show()
```

```python
# First, exclude the worst price outliers using a high quantile cutoff (e.g. keep the cheapest 99% of items) and replot:
import matplotlib.pyplot as plt

upper = amazon_df["price"].quantile(0.99)
filtered = amazon_df[amazon_df["price"] <= upper]

filtered["price"].hist(bins=50)
plt.xlabel("price")
plt.ylabel("frequency")
plt.show()
```

The current histogram is very hard to read for two reasons:
- there are items with extremely high prices >= 100000 Euros (outliers) that are forcing the chart to have a very long x axis compared to what's needed in order to show the majority of the product prices.
- Same happens with "frequency" where the vast majority of the product have a very similar and low pricing.

How to solve that:
- Exclude price related outliers
- Show the frequency bar in a way that it is reduced in relative size but still provide it's value.

```python
# and now I will apply a log scale on the y-axis so high frequencies shrink but stay comparable:
import matplotlib.pyplot as plt

upper = amazon_df["price"].quantile(0.99)
filtered = amazon_df[amazon_df["price"] <= upper]

ax = filtered["price"].hist(bins=50)
ax.set_yscale("log")
plt.xlabel("price")
plt.ylabel("frequency (log scale)")
plt.show()
```

### Part 3: Unpacking Product Ratings

Business Question: How do customers rate products on Amazon UK, and are there any patterns or tendencies in the ratings?

#### Measures of Centrality:

- Calculate the mean, median, and mode for the rating of products.
- How do customers generally rate products? Is there a common trend?

```python
mean_rating = amazon_df["stars"].mean()
median_rating = amazon_df["stars"].median()
mode_rating = amazon_df["stars"].mode()[0]

mean_rating, median_rating, mode_rating
```

We see that the median and mode rating of customers is 0, something expected as usually the people who want to write reviews are the people who are very dissatisfied with the product.

#### Measures of Dispersion:

- Determine the variance, standard deviation, and interquartile range for product rating.
- Are the ratings consistent, or is there a wide variation in customer feedback?

```python
var_rating = amazon_df["stars"].var()
std_rating = amazon_df["stars"].std()
inter_range_rating = amazon_df["stars"].quantile(0.75) - amazon_df["stars"].quantile(0.25)

var_rating, std_rating, inter_range_rating
```

For a scale of rating between 0 and 5, having a variance of 4.8 and a standard deviation of 2.19 constitute wide variation.

#### Shape of the Distribution:

- Calculate the skewness and kurtosis for the rating column.
- Are the ratings normally distributed, or do they lean towards higher or lower values?

```python
skew_stars = amazon_df["stars"].skew()
kurt_stars = amazon_df["stars"].kurt()

skew_stars, kurt_stars
```

Skew ≈ 0.08 indicates a distribution that is effectively symmetric.

Kurtosis ≈ −1.93 indicates that the tails are lighter and the peak is flatter than a normal distribution. This means extreme star ratings occur less frequently. 

#### Visualizations:

- Plot a histogram to visualize the distribution of product ratings. Is there a specific rating that is more common?

```python
import matplotlib.pyplot as plt

amazon_df["stars"].hist(bins=50)
plt.xlabel("stars")
plt.ylabel("frequency")
plt.show()
```

The rating 0 is the favorite rating of Amazon customers, expected as explained above.
